```python
import pandas as pd
import tkinter as tk
file1=pd.read_csv("D:\Ananya\studentdetails.csv")
size=list(file1.shape)
print(size)
file1
```

```python
def rollnum(x):
  roll_num_list=list(file1.at[x,'srn'])
  roll_num=int(roll_num_list[9])*1000+int(roll_num_list[10])*100+int(roll_num_list[11])*10+int(roll_num_list[12])
  return(roll_num)


def seat(x):
  room_no=int(x/60)
  seat_no=x%60
  return(room_no, seat_no)


def clear_frame():
    for widgets in frame3.winfo_children():
      widgets.destroy()
    entry1.delete(0,tk.END)
    entry2.delete(0,tk.END)

def Submit():
  name=entry1.get()
  srn=entry2.get()
  a=-1

  for i in file1['name']:
      a=a+1
      if i==name:
        if file1.at[a,'srn']==srn:

          label0=tk.Label(frame3,text="Name:",font=("Helvetica",15,"italic"))
          label0.grid(row=0,column=0)

          label1=tk.Label(frame3, text=name,font=("Helvetica",15,"italic"))
          label1.grid(row=0,column=3)

          label2=tk.Label(frame3, text="SRN:",font=("Helvetica",15,"italic"))
          label2.grid(row=1,column=0)

          label3=tk.Label(frame3, text=file1.at[a,'srn'],font=("Helvetica",15,"bold"))
          label3.grid(row=1,column=3)
          if file1.at[a,'attendance']>=85:
            if file1.at[a,'certificate']=='YES':

              label1=tk.Label(frame3, text="All requirements are met.",font=("Helvetica",15,"bold"))
              label1.grid(row=2,column=2)

              label2=tk.Label(frame3, text= "Course:",font=("Helvetica",15,"bold"))
              label2.grid(row=3,column=0)

              label3=tk.Label(frame3, text= file1.at[a,'course'],font=("Helvetica",15,"bold"))
              label3.grid(row=3,column=3)

              label4=tk.Label(frame3, text="Sem:",font=("Helvetica",15,"bold"))
              label4.grid(row=4,column=0)

              label5=tk.Label(frame3, text= file1.at[a,'sem'],font=("Helvetica",15,"bold"))
              label5.grid(row=4,column=3)

              label6=tk.Label(frame3, text= "Class:",font=("Helvetica",15,"bold"))
              label6.grid(row=5,column=0)

              label7=tk.Label(frame3, text= file1.at[a,'class'],font=("Helvetica",15,"bold"))
              label7.grid(row=5,column=3)

              label8=tk.Label(frame3, text="Rollno:",font=("Helvetica",15,"bold"))
              label8.grid(row=6,column=0)

              label9=tk.Label(frame3, text=rollnum(a),font=("Helvetica",15,"bold"))
              label9.grid(row=6,column=3)
              x,y=seat(rollnum(a))

              label10=tk.Label(frame3, text="Room Number",font=("Helvetica",15,"bold"))
              label10.grid(row=7,column=0)

              label11=tk.Label(frame3, text=x,font=("Helvetica",15,"bold"))
              label11.grid(row=7,column=3)

              label12=tk.Label(frame3, text="Seat Number:",font=("Helvetica",15,"bold"))
              label12.grid(row=8,column=0)

              label13=tk.Label(frame3, text=y,font=("Helvetica",15,"bold"))
              label13.grid(row=8,column=3)

              break
              
            else:
              label4=tk.Label(frame3, text="Permission denied. Certificates not submitted.",font=("Helvetica",15,"bold"))
              label4.grid(row=9,column=2)

              break

          else:
              label1=tk.Label(frame3, text="Permission denied. Attendance too low.",font=("Helvetica",15,"bold"))
              label1.grid(row=10,column=2)
          
              break
        else:
          label1=tk.Label(frame3, text="Permission denied srn is wrong",font=("Helvetica",15,"bold"))
          label1.grid()

      elif a==size[0]-1:
        label22=tk.Label(frame3, text="Error, not found",font=("Helvetica",15,"bold"))
        label22.grid(row=10,column=2)

root=tk.Tk()
root.title("PES Exam Admission Ticket")
root.geometry('690x600')

root.minsize(800, 800)


frame1 =tk.Frame(root, highlightbackground="black", highlightthickness=2,
                            height=50,width=100,padx=50,pady=30,bg='lavender')


#label
wish=tk.Label(master=frame1,text='Hall Ticket Form',
                  font=("Times New Roman",16,"bold"),
                  fg='black',
                  bg='lavender',
                  width=20,
                  height=1,padx=15,pady=15)
wish.grid(row=1,column=3)
    
#input
wish1=tk.Label(master=frame1,text='Name',font=("Calibri",14,"bold"),
               fg='black',
               bg='lavender',
               width=5,
               height=1,padx=5,pady=5)
wish1.grid(row=2,column=0)

entry1=tk.Entry(master=frame1)
entry1.grid(row=2,column=6)

#label
wish2=tk.Label(master=frame1,text='SRN',font=("Calibri",14,"bold"),
               fg='black',
               bg='lavender',
               width=5,
               height=1,padx=5,pady=5)
wish2.grid(row=4,column=0,padx=20,pady=20)

#input
entry2=tk.Entry(master=frame1)
entry2.grid(row=4,column=6,padx=20,pady=20)

#frames
frame2 =tk.Frame(root, highlightbackground="green", highlightthickness=2)
frame3=tk.Frame(root,height=50,width=50)

#submit
framebutton =tk.Button(frame2, text="Submit",bg='green',command=Submit,padx=10,pady=10,
                       height=2,width=5,font=("Calibri",14,"bold"))

framebutton.grid()


frame1.pack(padx=10, pady=5)
frame2.pack(padx=10, pady=5)
frame3.pack(padx=10,pady=5)

#to clear output
tk.Button(frame3, text="Clear", font=("Calibri",14,"bold"),command=clear_frame).grid(pady=20,row=11,column=5)

root.mainloop()
```
